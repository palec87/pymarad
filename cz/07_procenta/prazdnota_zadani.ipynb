Van der Waalsův poloměr popisuje vzdálenost, kdy sousední molekula pociťuje takovou odpudivou sílu (zejména mezi elektrony), že je takřka nemožné další přiblížení.
  * Představ si například odpuzující se magnety, nebo Zemi a slunce.
  * Jedná se sice o jiné síly, ale v obou případech na sebe objekty působí, ale také mají mezi sebou spoustu volného prostoru.
  * A Van der Waalsův objem tento volný prostor mezi molekulami zahrnuje, protože se na molekuly dívá z hlediska vzájemné interakce.
