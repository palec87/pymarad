## plot of the demand curve, which is transformed to bitcode.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = [6,4] # graph size
mpl.rcParams['lines.linewidth'] = 3  # line width
mpl.rcParams['lines.markersize'] = 10  # size of point markers
mpl.rcParams['xtick.labelsize'] = 13  # font size of x axis ticks
mpl.rcParams['ytick.labelsize'] = 13 # font size of y axis ticks
# mpl.rcParams['axes.labelsize'] = 'small'  # font size of the axes labels
```

```python
x = np.linspace(0, 100, 101)
a = -1/100
b = 1

plt.plot(x, a*x+b, lw=3)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.xlabel('Price per unit [money]', fontsize=12)
plt.ylabel('Demand for one unit [% fraction]', fontsize=12)
plt.title('Demand curve for a lemonade', fontsize=14)
plt.savefig('demand_lim.png', dpi=180)
plt.show()
```

```python
plt.plot(x, a*x+b, lw=3)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.xlabel('Cena za kus [peníz]', fontsize=12)
plt.ylabel('Poptávka po jenom kusu [% zlomek]', fontsize=12)
plt.title('Poptávka po limonádě', fontsize=14)
plt.savefig('poptavka_lim.png', dpi=180)
plt.show()
```

## 2D FFT of Mr Joseph Fourier himself.

```python
from PIL import Image,ImageOps
from scipy import fft
import numpy as np
import matplotlib.pyplot as plt

remmax = lambda x: x/x.max()
remmin = lambda x: x - np.amin(x, axis=(0,1), keepdims=True)
touint8 = lambda x: (remmax(remmin(x))*(256-1e-4)).astype(int)

import scipy.fftpack as fp

## Functions to go from image to frequency-image and back
im2freq = lambda data: fp.rfft(fp.rfft(data, axis=0),
                               axis=1)
freq2im = lambda f: fp.irfft(fp.irfft(f, axis=1),
                             axis=0)
```

```python
img = Image.open('Fourier2.jpg').convert('L')
# img = ImageOps.invert(img)
```

```python
img.save('fourier_greyscale.png')
```

```python
arr = im2freq(img)
# arr = np.absolute(fft.fft2(img))
# arr *= 255.0 / arr.max()
# arr = np.absolute(arr)
arr = touint8(arr)
arr
```

```python
img_fft = Image.fromarray(arr)
img_fft.save('fourier_fft.png')
img_fft.show()
```

```python
img2 = np.fft.fftshift(np.fft.fft2(img))
plt.figure(num=None, figsize=(8, 6), dpi=80)
# plt.imshow(abs(img2), cmap='gray')
plt.imshow(np.log(abs(img2)), cmap='gray')
plt.axis('off')
plt.savefig('fourier_fft.png',bbox_inches='tight')
```

```python
img2 *= 255.0 / img2.max()
img_fft = Image.fromarray(np.log(abs(img2)), mode='L')
img_fft.save('fourier_fft.png')
img_fft.show()
```
