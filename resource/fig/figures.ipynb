## plot of the demand curve, which is transformed to bitcode.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = [6,4] # graph size
mpl.rcParams['lines.linewidth'] = 3  # line width
mpl.rcParams['lines.markersize'] = 10  # size of point markers
mpl.rcParams['xtick.labelsize'] = 13  # font size of x axis ticks
mpl.rcParams['ytick.labelsize'] = 13 # font size of y axis ticks
# mpl.rcParams['axes.labelsize'] = 'small'  # font size of the axes labels
```

```python
x = np.linspace(0, 100, 101)
a = -1/100
b = 1

plt.plot(x, a*x+b, lw=3)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.xlabel('Price per unit [money]', fontsize=12)
plt.ylabel('Demand for one unit [% fraction]', fontsize=12)
plt.title('Demand curve for a lemonade', fontsize=14)
plt.savefig('demand_lim.png', dpi=180)
plt.show()
```

```python
plt.plot(x, a*x+b, lw=3)
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.xlabel('Cena za kus [peníz]', fontsize=12)
plt.ylabel('Poptávka po jenom kusu [% zlomek]', fontsize=12)
plt.title('Poptávka po limonádě', fontsize=14)
plt.savefig('poptavka_lim.png', dpi=180)
plt.show()
```
